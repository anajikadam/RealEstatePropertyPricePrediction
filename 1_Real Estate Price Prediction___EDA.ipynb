#### Fixing Numerical Features:

[Contents:](#Contents:)

##### Property Area:
- data type is string convert it to float
- Also in range (1200 to 1450) convert it to Average of that range
- Not digit present then set negative value

```python
numbers = re.compile(r"[-+]?(\d*\.\d+|\d+)") 
numbers
```

```python
numbers.findall(str('1200 to 1400'))
```

```python
numbers.findall(str('1200 +'))
```

```python
numbers.findall(str("shot/tea"))
```

```python
df['Property Area in Sq. Ft.'].value_counts().sample(5)
```

```python
numbers = re.compile(r"[-+]?(\d*\.\d+|\d+)")

def avg_property_area(x):
    x = numbers.findall(x)
    if len(x) == 1:                                # '1200 +'
        return np.float(x[0])
    elif len(x) == 2:                              # '1200 to 1400'
        return (np.float(x[0])+np.float(x[1]))/2   # return average of 2 numbers
    else:
        return -99                                 # else negative value return
    
  
df['Property Area in Sq. Ft. Cleaned'] = df['Property Area in Sq. Ft.'].apply(lambda x: avg_property_area(str(x)))
```
